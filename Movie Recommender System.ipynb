```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.shape
```

```python
credits.head()
```

```python
#merging these two datasets into one
movies=movies.merge(credits,on='title')
```

```python
movies.shape
```

```python
# Recommendation system types
# 1-  Content based (tags)
# 2-  collaborative filtering
# 3-  Hybrid

# We will make here content based recommendation system.We will recommend the 5 movies that has similar content like the user 
# watched.
```

```python
movies.head(1)
```

```python
#  we will only keep that columns in the dataset from which we can extract useful tags.
# --------genres,id,keywords,title,overview,tagline,cast,crew

df=movies[['movie_id','genres','keywords','overview','title','cast','crew']]
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
#Now we will perform processing on columns one by one
df['genres'][0]
```

```python
import ast
```

```python
def extract_tags(text):
    
    tags_list=[]
    for x in ast.literal_eval(text):
        tags_list.append(x['name'])
            
    return tags_list
```

```python
extract_tags(df['genres'][0])
```

```python
df.loc[:,'genres']=df['genres'].apply(extract_tags)
df.loc[:,'keywords']=df['keywords'].apply(extract_tags)
```

```python
def extract_top3cast(text):
    
    tag_list=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter <3:
            tag_list.append(i['name'])
        counter+=1
        
        
            
    return tag_list
        
```

```python
df['cast'][0]
```

```python
df.loc[:,'cast']=df['cast'].apply(extract_top3cast)
```

```python
df['cast'][0]
```

```python
def fetch_crew_director(text):
    dir_list=[]
    for  i in ast.literal_eval(text):
        if i['job']=='Director':
            dir_list.append(i['name'])
            
    return dir_list
```

```python
ast.literal_eval(df['crew'][0])
```

```python
df.loc[:,'crew']=df['crew'].apply(fetch_crew_director)
```

```python
df['crew'][1]
```

```python
df.head()
```

```python
df.loc[:,'overview']=df['overview'].apply(lambda x:x.split(' '))
```

```python
df.head()
```

```python
def collapse(text):
    l=[]
    
    for i in text:
        l.append(i.replace(' ',''))
        
    return l
```

```python
df.loc[:,'crew']=df['crew'].apply(collapse)
df.loc[:,'cast']=df['cast'].apply(collapse)
df.loc[:,'keywords']=df['keywords'].apply(collapse)
df.loc[:,'genres']=df['genres'].apply(collapse)
```

```python
df.head()
```

```python
df.loc[:,'tags']=df['genres']+df['keywords']+df['overview']+df['cast']+df['crew']
df.head()
```

```python
new_df=df.drop(columns=['genres','keywords','overview','cast','crew'])
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: ' '.join(x))
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
new_df['tags'][0]
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

ps=PorterStemmer()
stem_words=new_df['tags'].apply(lambda x: ' '.join([ps.stem(word) for word in word_tokenize(x)]))
```

```python
stem_words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english',max_features=5000)
count_vector=cv.fit_transform(stem_words).toarray()
```

```python
count_vector.shape
```

```python
count_vector[0]
```

```python
feature_names=cv.get_feature_names_out()
[print(x) for x in feature_names]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(count_vector)
similarity
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x: x[1])
```

```python
def recommend(movie):
    
    index=new_df[new_df['title']==movie].index[0]
    distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x: x[1])
    
    for i in distance[1:6]:
        print(new_df.iloc[i[0]]['title'])
    
```

```python
recommend('Avatar')
```
